```python
print("\nShor Code with BasicSimulator")
print("-----------------------------")

from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile
from qiskit.providers.basic_provider import BasicSimulator

# ---------------- Helper function for running circuits ----------------
def run_circuits(circuits):
    backend = BasicSimulator()
    transpiled_circuits = transpile(circuits, backend)
    job = backend.run(transpiled_circuits, shots=1)
    results = job.result()
    measured_bits = []
    for res in results.results:
        key = list(res.data.counts.keys())[0]  # "0x0" or "0x1"
        measured_bits.append(int(key, 16))     # convert hex to int
    return measured_bits

# ---------------- Single Qubit Example ----------------
single_circuits = []

qc = QuantumCircuit(1, 1)
qc.h(0)       # Prepare |+> state
qc.x(0)       # Bit flip error
qc.z(0)       # Phase flip error
qc.h(0)
qc.measure(0, 0)
single_circuits.append(qc)

single_results = run_circuits(single_circuits)
print("\nSingle Qubit with Bit Flip and Phase Error:")
print(single_results)


# ---------------- Shor Code (9 Qubits) ----------------
# Encode |0> into 9 qubits
shor_circuits = []

q = QuantumRegister(9, 'q')
c = ClassicalRegister(1, 'c')  # measure only logical qubit
shor_qc = QuantumCircuit(q, c)

# Encoding
shor_qc.cx(q[0], q[3])
shor_qc.cx(q[0], q[6])
shor_qc.h(q[0])
shor_qc.h(q[3])
shor_qc.h(q[6])
shor_qc.cx(q[0], q[1])
shor_qc.cx(q[3], q[4])
shor_qc.cx(q[6], q[7])
shor_qc.cx(q[0], q[2])
shor_qc.cx(q[3], q[5])
shor_qc.cx(q[6], q[8])
shor_qc.barrier()

# Introduce errors
shor_qc.x(q[0])
shor_qc.z(q[0])
shor_qc.barrier()

# Decode & correct bit flip
shor_qc.cx(q[0], q[1])
shor_qc.cx(q[3], q[4])
shor_qc.cx(q[6], q[7])
shor_qc.cx(q[0], q[2])
shor_qc.cx(q[3], q[5])
shor_qc.cx(q[6], q[8])
shor_qc.ccx(q[1], q[2], q[0])
shor_qc.ccx(q[4], q[5], q[3])
shor_qc.ccx(q[7], q[8], q[6])

# Decode & correct phase flip
shor_qc.h(q[0])
shor_qc.h(q[3])
shor_qc.h(q[6])
shor_qc.cx(q[0], q[3])
shor_qc.cx(q[0], q[6])
shor_qc.ccx(q[6], q[3], q[0])
shor_qc.barrier()

shor_qc.measure(q[0], c[0])
shor_circuits.append(shor_qc)

shor_results = run_circuits(shor_circuits)
print("\nShor Code Logical Qubit after Bit Flip & Phase Error Correction:")
print(shor_results)
```

```python
# ---------------- Exhaustive single–qubit fault-tolerance verification ----------------
# We inject X, Z, and Y on each of the 9 physical qubits and check if the logical qubit is recovered.

def shor_logical_test(error_qubit, error_type):
    q = QuantumRegister(9, 'q')
    c = ClassicalRegister(1, 'c')
    qc = QuantumCircuit(q, c)

    # ---- Encoding ----
    qc.cx(q[0], q[3])
    qc.cx(q[0], q[6])
    qc.h(q[0])
    qc.h(q[3])
    qc.h(q[6])
    qc.cx(q[0], q[1])
    qc.cx(q[3], q[4])
    qc.cx(q[6], q[7])
    qc.cx(q[0], q[2])
    qc.cx(q[3], q[5])
    qc.cx(q[6], q[8])
    qc.barrier()

    # ---- Inject single-qubit error ----
    if error_type == "X":
        qc.x(q[error_qubit])
    elif error_type == "Z":
        qc.z(q[error_qubit])
    elif error_type == "Y":
        qc.y(q[error_qubit])
    qc.barrier()

    # ---- Bit-flip correction ----
    qc.cx(q[0], q[1])
    qc.cx(q[0], q[2])
    qc.ccx(q[1], q[2], q[0])

    qc.cx(q[3], q[4])
    qc.cx(q[3], q[5])
    qc.ccx(q[4], q[5], q[3])

    qc.cx(q[6], q[7])
    qc.cx(q[6], q[8])
    qc.ccx(q[7], q[8], q[6])

    # ---- Phase-flip correction ----
    qc.h(q[0])
    qc.h(q[3])
    qc.h(q[6])
    qc.cx(q[0], q[3])
    qc.cx(q[0], q[6])
    qc.ccx(q[6], q[3], q[0])

    qc.measure(q[0], c[0])
    return qc


# Run all 27 cases: 9 qubits × {X, Z, Y}
test_circuits = []
labels = []

for qubit in range(9):
    for err in ["X", "Z", "Y"]:
        test_circuits.append(shor_logical_test(qubit, err))
        labels.append((qubit, err))

results = run_circuits(test_circuits)

print("\nExhaustive Single-Qubit Error Test (Shor Code):")
for i, res in enumerate(results):
    qid, etype = labels[i]
    status = "RECOVERED" if res == 0 else "FAILED"
    print(f"Error {etype} on qubit q[{qid}]  -> Logical measurement: {res}  -> {status}")
```

```python
# ---------------- Double–Qubit Error Test (Limit of Shor Code) ----------------
# The Shor code is distance-3: it corrects 1 error, but not guaranteed for 2.
# We now inject two simultaneous Pauli errors and show failures appear.

def shor_double_error_test(q1, q2, e1, e2):
    q = QuantumRegister(9, 'q')
    c = ClassicalRegister(1, 'c')
    qc = QuantumCircuit(q, c)

    # ---- Encoding ----
    qc.cx(q[0], q[3])
    qc.cx(q[0], q[6])
    qc.h(q[0])
    qc.h(q[3])
    qc.h(q[6])
    qc.cx(q[0], q[1])
    qc.cx(q[3], q[4])
    qc.cx(q[6], q[7])
    qc.cx(q[0], q[2])
    qc.cx(q[3], q[5])
    qc.cx(q[6], q[8])
    qc.barrier()

    # ---- Inject two errors ----
    for qubit, err in [(q1, e1), (q2, e2)]:
        if err == "X":
            qc.x(q[qubit])
        elif err == "Z":
            qc.z(q[qubit])
        elif err == "Y":
            qc.y(q[qubit])
    qc.barrier()

    # ---- Bit-flip correction ----
    qc.cx(q[0], q[1])
    qc.cx(q[0], q[2])
    qc.ccx(q[1], q[2], q[0])

    qc.cx(q[3], q[4])
    qc.cx(q[3], q[5])
    qc.ccx(q[4], q[5], q[3])

    qc.cx(q[6], q[7])
    qc.cx(q[6], q[8])
    qc.ccx(q[7], q[8], q[6])

    # ---- Phase-flip correction ----
    qc.h(q[0])
    qc.h(q[3])
    qc.h(q[6])
    qc.cx(q[0], q[3])
    qc.cx(q[0], q[6])
    qc.ccx(q[6], q[3], q[0])

    qc.measure(q[0], c[0])
    return qc


# Run a small set of representative double-error cases
double_tests = []
double_labels = []

test_cases = [
    (0, 1, "X", "X"),
    (0, 3, "Z", "Z"),
    (2, 4, "Y", "Y"),
    (1, 7, "X", "Z"),
    (5, 6, "Z", "X"),
    (3, 8, "Y", "X"),
]

for q1, q2, e1, e2 in test_cases:
    double_tests.append(shor_double_error_test(q1, q2, e1, e2))
    double_labels.append((q1, q2, e1, e2))

double_results = run_circuits(double_tests)

print("\nDouble-Qubit Error Test (Shor Code Limits):")
for i, res in enumerate(double_results):
    q1, q2, e1, e2 = double_labels[i]
    status = "RECOVERED" if res == 0 else "FAILED"
    print(f"Errors {e1} on q[{q1}] and {e2} on q[{q2}] -> Logical measurement: {res} -> {status}")
```

```python
def run_aer_circuits(circuits, shots=1024):
    backend = AerSimulator()
    tcircuits = transpile(circuits, backend)
    measured = []
    for qc in tcircuits:
        job = backend.run(qc, shots=shots)
        result = job.result()
        counts = result.get_counts()
        # Take the most frequent outcome
        key = max(counts, key=counts.get)
        measured.append(int(key, 2))
    return measured

# -------------------------------
# Shor Encode 9-qubit logical 0
# -------------------------------
def shor_encode(qc, q):
    for i in [0, 3, 6]:
        qc.h(q[i])
        qc.cx(q[i], q[i+1])
        qc.cx(q[i], q[i+2])

# -------------------------------
# Simplified error correction
# -------------------------------
def shor_correct(qc, q):
    # Bit-flip correction (majority vote)
    for i in [0, 3, 6]:
        qc.ccx(q[i+1], q[i+2], q[i])

    # Phase-flip correction
    for i in [0, 3, 6]:
        qc.h(q[i])
    qc.ccx(q[3], q[6], q[0])
    for i in [0, 3, 6]:
        qc.h(q[i])

# -------------------------------
# Logical X
# -------------------------------
def logical_X_shor(qc, q):
    qc.x(q[0])
    qc.x(q[3])
    qc.x(q[6])

# -------------------------------
# Logical X test
# -------------------------------
def shor_logical_X_test():
    q = QuantumRegister(9, "q")
    c = ClassicalRegister(1, "c")
    qc = QuantumCircuit(q, c)

    shor_encode(qc, q)
    qc.barrier()
    logical_X_shor(qc, q)
    qc.barrier()
    shor_correct(qc, q)
    qc.barrier()
    qc.measure(q[0], c[0])
    return qc

# -------------------------------
# Single-qubit error test
# -------------------------------
def shor_single_error_test(error_qubit, error_type):
    q = QuantumRegister(9, "q")
    c = ClassicalRegister(1, "c")
    qc = QuantumCircuit(q, c)

    shor_encode(qc, q)
    qc.barrier()

    if error_type == "X":
        qc.x(q[error_qubit])
    elif error_type == "Z":
        qc.z(q[error_qubit])
    elif error_type == "Y":
        qc.y(q[error_qubit])
    qc.barrier()

    shor_correct(qc, q)
    qc.barrier()
    qc.measure(q[0], c[0])
    return qc

# -------------------------------
# Run tests
# -------------------------------
# Logical X
res = run_aer_circuits([shor_logical_X_test()])
print("Logical X Test | Expected: 1 | Obtained:", res[0])

# Exhaustive single-qubit errors
circuits = []
labels = []
for qubit in range(9):
    for err in ["X", "Z", "Y"]:
        circuits.append(shor_single_error_test(qubit, err))
        labels.append((qubit, err))

results = run_aer_circuits(circuits)
print("\nSingle-Qubit Error Test:")
for i, r in enumerate(results):
    qid, et = labels[i]
    status = "RECOVERED" if r == 0 else "FAILED"
    print(f"Error {et} on q[{qid}] -> Logical measurement: {r} -> {status}")
```
